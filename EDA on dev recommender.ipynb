```python
def print_description(i):
    for s in df['description']:
        print("-------------------------")
        print(s)
        i -= 1
        if i == 0:
            break
```

```python
print_description(10)
```

## Remove digits, punctuation, extra white spaces

```python
import string
df['description']=df['description'].apply(lambda x: re.sub('\w*\d\w*','', x))
df['description']=df['description'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
df['description']=df['description'].apply(lambda x: re.sub(' +',' ',x))
```

```python
print_description(10)
```

## Lowercase the description

```python
df['description']=df['description'].apply(lambda x: x.lower())
print_description(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(analyzer='word')
data=cv.fit_transform(df['description'])
df_dtm = pd.DataFrame(data.toarray(), columns=cv.get_feature_names())
df_dtm.index=df.index
df_dtm.head(3)
```

```python
df.index
```

```python
merged_df = df_dtm.merge(df['owner'], left_index=True, right_index=True, how='left')
merged_df.head()
```
